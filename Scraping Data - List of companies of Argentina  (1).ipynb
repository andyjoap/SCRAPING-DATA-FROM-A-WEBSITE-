```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Argentina'

page = requests.get(url)
print(page)
```

```python
soup = BeautifulSoup(page.text,'html')
print(soup)
```

```python
table1 = soup.find_all('table')[0]
print(table1)
```

```python
world_titles = table1.find_all('tr')[0]
print(world_titles)
```

```python
world_titles1 = table1.find_all('th')
print(world_titles1)
```

```python
world_titles2 = [title.text.strip() for title in world_titles1][:7]


world_titles2.insert(8, 'Status')
print(world_titles2)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_titles2)
df
```

```python
#Now we want to obtain all the data
column_data = table1.find_all('tr')
print(column_data)
```

```python
for row in column_data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    #print(individual_row_data)

    length = len(df)
    df.loc[length] = individual_row_data
    
```
